```python
import math
from matplotlib import pyplot as plt
```

## Scalar product

```python
def s(n):
    return (2*n - 1) / (1 + math.log(n, 2))
```

```python
xs = list(range(1, 1001))
ys = [s(n) for n in xs]
```

```python
plt.figure()
plt.plot(xs, ys, 'r')
plt.xlabel('n')
plt.ylabel('S(n)')
plt.show()
```

```python
def e(n):
    return s(n) / n
```

```python
ys = [e(n) for n in xs]
```

```python
plt.figure()
plt.plot(xs, ys, 'g')
plt.xlabel('n')
plt.ylabel('E(n)')
plt.show()
```

```python
def s2(n, p):
    return (2*n - 1) / ((n / p) + (n / p - 1) + math.log(p, 2))
```

```python
ys = [s2(n, 4) for n in xs]
```

```python
plt.figure()
plt.plot(xs, ys, 'r')
plt.xlabel('n')
plt.ylabel('S(n)')
plt.show()
```

```python
def e2(n, p):
    return s2(n, p) / p
```

```python
ys = [e2(n, 4) for n in xs]
```

```python
plt.figure()
plt.plot(xs, ys, 'g')
plt.xlabel('n')
plt.ylabel('E(n)')
plt.show()
```

## Matrix multiplication

```python
def s(n):
    return (2 * n*n*n - n*n) / (1 + math.log(n, 2))
```

```python
xs = list(range(1, 21))
ys = [s(n) for n in xs]
```

```python
plt.figure()
plt.plot(xs, ys, 'r')
plt.xlabel('n')
plt.ylabel('S(n)')
plt.show()
```
