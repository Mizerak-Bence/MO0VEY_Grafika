```python
import os

import yaml
```

```python
DIR_PATH = 'exercises'
```

```python
latex_head = r"""
\documentclass[a4paper]{article}

% Set margins
\usepackage[hmargin=2.5cm, vmargin=3cm]{geometry}

\frenchspacing

% Language packages
\usepackage[utf8]{inputenc}
\usepackage[T1]{fontenc}
\usepackage[magyar]{babel}

% AMS
\usepackage{amssymb,amsmath}

\usepackage{xcolor}

\begin{document}

\pagestyle{empty}
"""

latex_body = r"""
\noindent \textbf{ {{index}}.}

{{exercise}}

\bigskip

\noindent {\color{blue} {{solution}} }

\bigskip
"""

latex_tail = r"""
\end{document}
"""
```

```python
with open('/tmp/all_exercises.tex', 'w') as tex_file:
    tex_file.write(latex_head)
    for name in os.listdir(DIR_PATH):
        exercise_collection_path = os.path.join(DIR_PATH, name)
        with open(exercise_collection_path) as f:
            exercises = yaml.load(f, yaml.Loader)['exercises']
            for index, exercise in enumerate(exercises):
                body = latex_body
                body = body.replace('{{index}}', name.replace('_', '\\_') + '/' + str(index))
                body = body.replace('{{exercise}}', exercise['question'])
                body = body.replace('{{solution}}', str(exercise['answer']))
                tex_file.write(body)
    tex_file.write(latex_tail)
```
